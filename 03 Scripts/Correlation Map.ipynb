## 4. Regression Dataframe

```python
# Rename variable
df_cwur.rename(columns = {
	'institution': 'university',
    'score': 'total_score'}, inplace = True)
df_cwur
```

```python
# Add a new column to represent the institution
df_arwu['institution'] = 'ARWU'
df_cwur['institution'] = 'CWUR'
df_times['institution'] = 'Times'
```

```python
df_arwu['institution'].value_counts(dropna = False)
```

```python
df_cwur['institution'].value_counts(dropna = False)
```

```python
df_times['institution'].value_counts(dropna = False)
```

```python
# Keep relevant variables
df_reg_arwu = df_arwu[['total_score', 'institution']].copy()
df_reg_cwur = df_cwur[['total_score', 'institution']].copy()
df_reg_times = df_times[['total_score', 'institution']].copy()
df_reg_arwu
```

```python
df_reg_cwur
```

```python
df_reg_times
```

```python
# Merge
df_reg_combined = pd.concat([df_reg_arwu, df_reg_cwur, df_reg_times], ignore_index=True)
df_reg_combined
```

```python
df_reg_combined['institution'].value_counts(dropna = False)
```

```python
# Dummy variables
df_reg_combined = pd.get_dummies(df_reg_combined, columns=['institution'], drop_first=True)
df_reg_combined
```

ARWU is the baseline for the regression analysis.

```python
# Convert T/F
dummy_columns = ['institution_CWUR', 'institution_Times']
df_reg_combined[dummy_columns] = df_reg_combined[dummy_columns].astype(int)
df_reg_combined
```

# 3. Regression Analysis

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Set regression model
X = df_reg_combined[['institution_CWUR', 'institution_Times']]
y = df_reg_combined['total_score']
X = sm.add_constant(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = sm.OLS(y_train, X_train).fit()
print(model.summary())
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Create scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, color='red')
plt.xlabel('Actual Total Score')
plt.ylabel('Predicted Total Score')
plt.title('Actual vs Predicted Total Score')
plt.savefig('actual_vs_predicted_total_score.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.savefig('actual_vs_predicted_total_score.png', dpi=300, bbox_inches='tight')
```

```python
# Create scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(y_test[df_reg_combined['institution_CWUR'] == 1], 
            y_pred[df_reg_combined['institution_CWUR'] == 1], 
            color='blue', alpha=0.5, label='CWUR')
plt.scatter(y_test[df_reg_combined['institution_Times'] == 1], 
            y_pred[df_reg_combined['institution_Times'] == 1], 
            color='green', alpha=0.5, label='Times')
plt.scatter(y_test[(df_reg_combined['institution_CWUR'] == 0) & 
                   (df_reg_combined['institution_Times'] == 0)], 
            y_pred[(df_reg_combined['institution_CWUR'] == 0) & 
                   (df_reg_combined['institution_Times'] == 0)], 
            color='orange', alpha=0.5, label='ARWU')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, color='red')
plt.xlabel('Actual Total Score')
plt.ylabel('Predicted Total Score')
plt.title('Actual vs Predicted Total Score')
plt.legend(title='Institution')
plt.savefig('actual_vs_predicted_total_score_by_institution.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Assign regions based on the country
def assign_region(country):
    if country in ['China', 'Hong Kong', 'Taiwan', 'Singapore', 'South Korea', 'Japan']:
        return 'China-Aligned Region'
    elif country in ['United States of America', 'United Kingdom', 'Canada', 'Australia', 'Netherlands', 'Germany', 'France', 'Sweden', 'Switzerland', 'Belgium', 'Denmark', 'Norway', 'Finland', 'Israel']:
        return 'Western-Aligned Region'
    else:
        return 'Non-Aligned'
df_combined['alignment'] = df_combined['country/region'].apply(assign_region)
df_combined
```
