# 3. Time Series Analysis

```python
# Decompose the time series using an additive model
decomposition = sm.tsa.seasonal_decompose(df_line_population, model='additive')
```

```python
# Plot the decomposition
from pylab import rcParams
rcParams['figure.figsize'] = 18, 7
decomposition.plot()
plt.show()
```

The data didn't have any noise, so the level and trend appear to be the same. The dominant trend is gradual increase over time. Because the trend is strong and consistent, the seasonality and residual curves are flat which means there is no seasonality.

```python
from statsmodels.tsa.stattools import adfuller
def dickey_fuller(timeseries):
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the Dickey-Fuller test
dickey_fuller(df_line_population['country_population_mil'])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_line_population)
plt.show()
```

The high test statistic higher than the critical values and a p-value well above .05 (which fails to reject the null hypothesis - the series is not stationary) indicate that the time series is not stationary.

```python
# Stationarize data
data_diff = df_line_population - df_line_population.shift(1)
data_diff
```

```python
# Plot the data
plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
```
