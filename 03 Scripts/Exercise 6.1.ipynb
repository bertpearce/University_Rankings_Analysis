```python
# Replace NaN values with empty strings
df_arwu['university_name'] = df_arwu['university_name'].fillna('')
```

```python
from fuzzywuzzy import process

# Function to get the best match
def get_best_match(name, choices):
    best_match = process.extractOne(name, choices, scorer=process.fuzz.token_sort_ratio)
    return best_match

# Apply fuzzy matching to find the best match for each university name in df_arwu
df_arwu['best_match'] = df_arwu['university_name'].apply(lambda x: get_best_match(x, df_school['university_name'].tolist()))

# Extract the match name and score
df_arwu['match_name'] = df_arwu['best_match'].apply(lambda x: x[0] if x else None)
df_arwu['match_score'] = df_arwu['best_match'].apply(lambda x: x[1] if x else None)

# Identify unmatched names (with score below a threshold, e.g., 100)
unmatched = df_arwu[df_arwu['match_score'] < 100]

print("Unmatched University Names:")
print(unmatched[['university_name', 'match_name', 'match_score']])
```

```python
unmatched
```

```python
# Update the university_name column with the matched names
df_arwu['university_name'] = df_arwu['match_name']

# Drop the intermediate columns
df_arwu = df_arwu.drop(columns=['best_match', 'match_name', 'match_score'])

print("\nUpdated DataFrame with matched university names:")
print(df_arwu)
```
